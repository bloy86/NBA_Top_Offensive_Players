```python
import pandas as pd
import numpy as np

#File to load
file = "Seasons_stats.csv"

#Reading file and storing to dataframe
nba_data = pd.read_csv(file)
nba_data.head()
```

```python
#Filtering data to specific years and grouping by player
player_data = nba_data.loc[nba_data['Year'] >= 1980]
player_data = player_data.groupby(['Player'])

player_data.head()
```

```python
#Offensive Analysis
#Creating specific offense data frame
offense_data = player_data[['Year', 'Player', 'Pos', 'G', 'MP', 'TS%', 'eFG%', 'AST', 'PTS', 'ORB']]
offense_data.head()
```
