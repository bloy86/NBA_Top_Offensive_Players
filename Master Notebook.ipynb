```python
#Adding normalizing columns
#Normalizing Data: (Total X - Min X) / (Max X - Min X)
df_2018['PTS_Norm'] = (df_2018['PTS'] - df_2018['PTS'].min()) / (df_2018['PTS'].max() - df_2018['PTS'].min())
df_2018.sort_values('PTS_Norm', ascending=False).head(10)
```

```python
#Defining function to add normalizing columns to all key stats
def normalize(col):
    return (col - col.min()) / (col.max() - col.min())
```

```python
cols_to_norm = [
    'PTS',
    'MP',
    'FG',
    'FGA',
    '3P',
    '3PA',
    'FT',
    'FTA',
    'ORB',
    'DRB',
    'AST',
    'STL',
    'TOV',
    'BLK',
]
```

```python
def vorp(df):
    for col_name in cols_to_norm:
        df['{}_norm'.format(col_name)] = normalize(df(col_name))
    return df
```

```python
df_2018_norm = vorp(df_2018)
```
