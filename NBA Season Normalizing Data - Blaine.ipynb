# Normalizing Player Stats by Season

```python
#Normalizing Data Across Seasons
#Someone who averages 22PPG in 1980 is different than someone who averaged 22PPG in 2017
```

```python
#Filtering on 2017 to get a smaller dataframe to work with
season = (df_filter['Year'] == 2017)
df_2017 = df_filter[season]
df_2017
```

```python
#Adding normalizing columns to smaler 2017 data frame
#Normalizing Data: (Total X - Min X) / (Max X - Min X)
df_2017['PTS_Norm'] = (df_2017['PTS'+] - df_2017['PTS'].min()) / (df_2017['PTS'].max() - df_2017['PTS'].min())
df_2017.sort_values('PTS_Norm', ascending=False).head(10)
```

```python
#Defining function to add normalizing columns to all key stats
def normalize(col):
    return (col - col.min()) / (col.max() - col.min())
```

```python
#Looking at all important stat columns we want to normalize
cols_to_norm = [
    'PTS',
    'MP',
    'FG',
    'FGA',
    '3P',
    '3PA',
    'FT',
    'FTA',
    'ORB',
    'DRB',
    'AST',
    'STL',
    'TOV',
    'BLK',
]
```

```python
#Defining vorp function (Value over Replacement Player) for the dataframe that will loop through the specific columns above and add new column with normalized data
def vorp(df):
    for col_name in cols_to_norm:
        df['{}_norm'.format(col_name)] = normalize(df[col_name])
    return df
```

```python
#Looking at specific 2017 dataframe and testing vorp function
df_2017_norm = vorp(df_2017)
df_2017_norm.head()
```

```python
#Grouping all years together into new data frame to run the vorp function
grouped = df_filter.groupby('Year')
grouped['PTS'].mean()
```

# Calculating Normalized Stat Volume Variance between similar players

```python
#Creating Distance Function that allows players to be compared against eachother for multiple stats
def calc_distance(u, v):
    dist = np.sqrt(np.sum((u - v)**2))
    return dist
```
